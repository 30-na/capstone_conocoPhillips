```python
import math
def r_to_z(r):
    return math.log((1 + r) / (1 - r)) / 2.0

def z_to_r(z):
    e = math.exp(2 * z)
    return((e - 1) / (e + 1))

def r_confidence_interval(r, alpha, n):
    z = r_to_z(r)
    se = 1.0 / math.sqrt(n - 3)
    z_crit = stats.norm.ppf(1 - alpha/2)  # 2-tailed z critical value

    lo = z - z_crit * se
    hi = z + z_crit * se

    # Return a sequence
    return (z_to_r(lo), z_to_r(hi))
```

```python
# 1. Use linear regression, train on known data, 'test' on NA. 
# X = cumulative data, y=peak 12 month, then do X=12 peak, y= 9 month, etc., etc. Use output of linear regression prediction to replace unknown y

# To replace: Build model after using dropna() on columns of interest. use np.where(isnull) 
# to pull NA datapoints as your "test" dataset. To replace, use np.where(isna) replace with new prediction.
```

```python
Cum_to_month12 = data_adj[['12 Month Cum. Production (BBL)', 'Peak Rate 12 Month (BOPD)']]
#Pull Columns
CompleteData = Cum_to_month12.dropna()
NAData = Cum_to_month12[Cum_to_month12.isna().any(axis=1)]

#Train
CumProdTrain = CompleteData['12 Month Cum. Production (BBL)'].values.reshape(-1,1)
Peak12Train = CompleteData['Peak Rate 12 Month (BOPD)'].values.reshape(-1,1)
Reg1 = LinReg().fit(CumProdTrain,Peak12Train)

#Predict
CumProdTest = NAData['12 Month Cum. Production (BBL)'].values.reshape(-1,1)
Peak12Prediction = Reg1.predict(CumProdTest)

#Replace
data_adj.loc[data_adj['Peak Rate 12 Month (BOPD)'].isna(), 'Peak Rate 12 Month (BOPD)'] = Peak12Prediction
```

### Random Forest

```python
#RF with Controllable Features as predictors and production as response
data_adj=data_adj.drop(columns=['Fluid Type from DI'])
random_forest_data = data_adj.dropna(axis=0)
results=random_forest_data
Controllable = random_forest_data[['Formation','Stage Length (ft/stage)','Fracture Design (LBS/GAL)'
                                   ,'Completion Type','Well Spacing (ft)','Latitude','Longitude','Lateral Length (ft)','Stages','Fluid (BBL)','Total Prop (LBS)']]
```

```python
data_adj.isnull().sum()
```

```python
random_forest_data.shape
```

```python
enc = OrdinalEncoder()
```

```python
Categorical_Controllable = Controllable[['Formation','Completion Type']]
Categorical_Controllable_Enc = enc.fit_transform(Categorical_Controllable)
form_coded = Categorical_Controllable_Enc[:,0]
comptype_coded = Categorical_Controllable_Enc[:,1]

X = Controllable
X['Formation'] = form_coded
X['Completion Type'] = comptype_coded
```

```python
Column_Names = X.columns
```

```python
#Scaling Data
Continuous_Controllable = Controllable[['Stage Length (ft/stage)','Fracture Design (LBS/GAL)','Well Spacing (ft)'
                                        ,'Lateral Length (ft)','Stages','Fluid (BBL)','Total Prop (LBS)']]
scaler=mmscale()
Continous_Transformed = scaler.fit_transform(Continuous_Controllable)
stagelength_scaled = Continous_Transformed[:,0] 
fracdesign_scaled = Continous_Transformed[:,1] 
spacing_scaled = Continous_Transformed[:,2]
latlength_scaled = Continous_Transformed[:,3] 
stages_scaled = Continous_Transformed[:,4]
fluid_scaled = Continous_Transformed[:,5] 
prop_scaled = Continous_Transformed[:,6] 
X['Stage Length (ft/stage)'] = stagelength_scaled
X['Fracture Design (LBS/GAL)'] = fracdesign_scaled
X['Well Spacing (ft)'] = spacing_scaled
X['Lateral Length (ft)'] = latlength_scaled
X['Stages'] = stages_scaled
X['Fluid (BBL)'] = fluid_scaled
X['Total Prop (LBS)'] = prop_scaled 
```

```python
X.shape
```

```python
param_grid = {'max_depth': [2, 5, 7, 10, 12, 15],'n_estimators': [200,500,600,700,800,1000]}
```

```python
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
from sklearn import metrics   #Additional scklearn functions
from sklearn.model_selection import GridSearchCV
```

```python
def modelfit(alg, dtrain, dtest,useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(dtrain.values, label=dtest.values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='auc', early_stopping_rounds=early_stopping_rounds)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    alg.fit(dtrain, dtest,eval_metric='auc')
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain)
    
    
    #Print model report:
    print("\nModel Report")
    print ("Accuracy : %.4g" % metrics.r2_score(dtest.values, dtrain_predictions))
    
                    
  #  feat_imp = pd.Series(alg.booster().get_fscore()).sort_values(ascending=False)
  #  feat_imp.plot(kind='bar', title='Feature Importances')
  #  plt.ylabel('Feature Importance Score')
```
